```python
#Countdown Timer app
import time
from tqdm import tqdm
```

```python
# Step 4: Get User Input
# Prompt the user to enter the time in minutes and seconds
minutes = int(input("Enter the time in minutes:"))
seconds = int(input("Enter the time in seconds:"))

# Convert the total time into seconds
total_time = minutes * 60 + seconds # Initialize the progress bar with the total number of seconds

for remaining_seconds in tqdm(range(total_time, 0, -1), desc="Countdown", unit="s", ncols=80):

# Calculate the remaining minutes and seconds
    mins, secs = divmod(remaining_seconds, 60)

    # Print the time in minutes and seconds
    print("{:02d}:{:02d}".format(minutes, seconds), end="\r")
    # Wait for one second
    time.sleep(1)
# Print a message when the time reaches zero
print("Time's up!")
```
